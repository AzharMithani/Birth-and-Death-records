### 1. Insights on Daily information

```python
day_wise = pd.read_csv('Insights/Day_wise_information.csv',header = 0)
day_wise['DOE']= pd.to_datetime((day_wise.Year*10000+day_wise.Month*100+day_wise.Day).apply(str),format='%Y%m%d')
daily_insights = pd.DataFrame(columns=['male_births','female_births','total_births','male_deaths','female_deaths','total_deaths'],index=['Mean','Maximum','Max_Date','Minimum','Min_Date'])
insights, temp = [], []

avg_male_births = day_wise.male_births.mean()
insights.append("Average Male Births per day = " + str(int(avg_male_births)))
temp.append(int(avg_male_births))

max_male_birth_day = day_wise.male_births.idxmax()
b = int(day_wise.male_births[[max_male_birth_day]])
d = str(day_wise.DOE[[max_male_birth_day]])[6:]
insights.append(("Maximum Male Births of '"+ str(b) + "' registered on date: "+ str(d))[:59])
temp.append(b)
temp.append(str(d)[:10])

min_male_birth_day = day_wise.male_births.idxmin()
b = int(day_wise.male_births[[min_male_birth_day]])
d = str(day_wise.DOE[[min_male_birth_day]])[6:]
insights.append(("Minimum Male Births of '"+ str(b) + "' registered on date: "+ str(d))[:59])
temp.append(b)
temp.append(str(d)[:11])

daily_insights.male_births = temp
temp = []

avg_female_births = day_wise.female_births.mean()
insights.append("Average Female Births per day = " + str(int(avg_female_births)))
temp.append(int(avg_female_births))

max_female_birth_day = day_wise.female_births.idxmax()
b = int(day_wise.female_births[[max_female_birth_day]])
d = str(day_wise.DOE[[max_female_birth_day]])[6:]
insights.append(("Maximum Female Births of '"+ str(b) + "' registered on date: "+ str(d))[:62])
temp.append(b)
temp.append(str(d)[:11])

min_female_birth_day = day_wise.female_births.idxmin()
b = int(day_wise.female_births[[min_female_birth_day]])
d = str(day_wise.DOE[[min_female_birth_day]])[6:]
insights.append(("Minimum Female Births of '"+ str(b) + "' registered on date: "+ str(d))[:61])
temp.append(b)
temp.append(str(d)[:11])

daily_insights.female_births = temp
temp = []

avg_total_births = day_wise.total_births.mean()
insights.append("Average Total births per day = " + str(int(avg_total_births)))
temp.append(int(avg_total_births))

max_total_birth_day = day_wise.total_births.idxmax()
b = int(day_wise.total_births[[max_total_birth_day]])
d = str(day_wise.DOE[[max_total_birth_day]])[6:]
insights.append(("Maximum Total births of '"+ str(b) + "' registered on date: "+ str(d))[:60])
temp.append(b)
temp.append(str(d)[:10])

min_total_birth_day = day_wise.total_births.idxmin()
b = int(day_wise.total_births[[min_total_birth_day]])
d = str(day_wise.DOE[[min_total_birth_day]])[6:]
insights.append(("Minimum Total births of '"+ str(b) + "' registered on date: "+ str(d))[:60])
temp.append(b)
temp.append(str(d)[:11])

daily_insights.total_births = temp
temp = []

for i in insights:
    all_insights.append(i)
    
insights = []

avg_male_deaths = day_wise.male_deaths.mean()
insights.append("Average Male Deaths per day = " + str(int(avg_male_deaths)))
temp.append(int(avg_male_deaths))

max_male_death_day = day_wise.male_deaths.idxmax()
b = int(day_wise.male_deaths[[max_male_death_day]])
d = str(day_wise.DOE[[max_male_death_day]])[6:]
insights.append(("Maximum Male deaths of '"+ str(b) + "' registered on date: "+ str(d))[:58])
temp.append(b)
temp.append(str(d)[:10])

min_male_death_day = day_wise.male_deaths.idxmin()
b = int(day_wise.male_deaths[[min_male_death_day]])
d = str(day_wise.DOE[[min_male_death_day]])[6:]
insights.append(("Minimum Male deaths of '"+ str(b) + "' registered on date: "+ str(d))[:58])
temp.append(b)
temp.append(str(d)[:11])

daily_insights.male_deaths = temp
temp = []

avg_female_deaths = day_wise.female_deaths.mean()
insights.append("Average Female Deaths per day = " + str(int(avg_female_deaths)))
temp.append(int(avg_female_deaths))

max_female_death_day = day_wise.female_deaths.idxmax()
b = int(day_wise.female_deaths[[max_female_death_day]])
d = str(day_wise.DOE[[max_female_death_day]])[6:]
insights.append(("Maximum Female deaths of '"+ str(b) + "' registered on date: "+ str(d))[:61])
temp.append(b)
temp.append(str(d)[:11])

min_female_death_day = day_wise.female_deaths.idxmin()
b = int(day_wise.female_deaths[[min_female_death_day]])
d = str(day_wise.DOE[[min_female_death_day]])[6:]
insights.append(("Minimum Female deaths of '"+ str(b) + "' registered on date: "+ str(d))[:60])
temp.append(b)
temp.append(str(d)[:11])

daily_insights.female_deaths = temp
temp = []

avg_total_deaths = day_wise.total_deaths.mean()
insights.append("Average Total Deaths per day = " + str(int(avg_total_deaths)))
temp.append(int(avg_total_deaths))

max_total_death_day = day_wise.total_deaths.idxmax()
b = int(day_wise.total_deaths[[max_total_death_day]])
d = str(day_wise.DOE[[max_total_death_day]])[6:]
insights.append(("Maximum Total deaths of '"+ str(b) + "' registered on date: "+ str(d))[:61])
temp.append(b)
temp.append(str(d)[:11])

min_total_death_day = day_wise.total_deaths.idxmin()
b = int(day_wise.total_deaths[[min_total_death_day]])
d = str(day_wise.DOE[[min_total_death_day]])[6:]
insights.append(("Minimum Total deaths of '"+ str(b) + "' registered on date: "+ str(d))[:60])
temp.append(b)
temp.append(str(d)[:11])

daily_insights.total_deaths = temp
temp = []

for i in insights:
    all_insights.append(i)
    
display(daily_insights)
daily_insights['Data'] = daily_insights.index.values
daily_insights = daily_insights[['Data','male_births','female_births','total_births','male_deaths','female_deaths','total_deaths']]
daily_insights.to_csv('Insights/Daily_Insights.csv',index = None)
```

### 3. Birth and Death Records not registered in Detected cases and deaths due to diseases
   The table below shows the difference between the number of registered births and deaths between 2 datasets "Birth and death
   records" and "Detected cases and deaths due to diseases" from year 2014 to 2017

```python
sup_data = pd.read_csv("Detected_cases_and_deaths.csv")
our_data = pd.read_csv("Insights/Year_wise_information.csv",index_col=None)
our_data['total_births'] = our_data.male_births + our_data.female_births
our_data['total_deaths'] = our_data.male_deaths + our_data.female_deaths
our_data = our_data[['Year','male_births','female_births','total_births','male_deaths','female_deaths','total_deaths']]
our_data = our_data.drop([0,1,2,3,8],axis=0)

diff_data = pd.DataFrame(columns=['Year','male_births','female_births','total_births','male_deaths','female_deaths','total_deaths'])
diff_data.male_births = our_data.male_births.values - sup_data.male_births.values
diff_data.female_births = our_data.female_births.values - sup_data.female_births.values
diff_data.total_births = our_data.total_births.values - sup_data.total_births.values
diff_data.male_deaths = our_data.male_deaths.values - sup_data.male_deaths.values
diff_data.female_deaths = our_data.female_deaths.values - sup_data.female_deaths.values
diff_data.total_deaths = our_data.total_deaths.values - sup_data.total_deaths.values
diff_data.Year = our_data.Year.values



display(diff_data)
diff_data.to_csv('Insights/Recording_differences_between_datasets.csv',index = None)
```
